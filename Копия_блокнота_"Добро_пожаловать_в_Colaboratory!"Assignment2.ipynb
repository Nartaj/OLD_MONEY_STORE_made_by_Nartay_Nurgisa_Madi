```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from google.colab import files

uploaded = files.upload()
df = pd.read_excel(next(iter(uploaded)))

df.drop(['Name', 'User_ID'], axis=1, inplace=True)

df['Last_Login'] = pd.to_datetime(df['Last_Login'], errors='coerce')
df['days_since_login'] = (pd.Timestamp.now() - df['Last_Login']).dt.days
df.drop(['Last_Login'], axis=1, inplace=True)


df['Subscription_Type'] = df['Subscription_Type'].astype('category').cat.codes
df['Favorite_Genre'] = df['Favorite_Genre'].astype('category').cat.codes
df['Country'] = df['Country'].astype('category').cat.codes

df['Watch_Time_Hours'] = pd.to_numeric(df['Watch_Time_Hours'], errors='coerce')

df.dropna(subset=['Watch_Time_Hours'], inplace=True)

df['heavy_user'] = df['Watch_Time_Hours'].apply(lambda x: 1 if x > 20 else 0)

X = df.drop('heavy_user', axis=1)
y = df['heavy_user']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

base_model = DecisionTreeClassifier(random_state=42)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)

print("📊 До тюнинга:")
print("Accuracy:", accuracy_score(y_test, y_pred_base))
print("F1-score:", f1_score(y_test, y_pred_base, average='weighted'))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_base))
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
best_dt = grid.best_estimator_
y_pred_tuned = best_dt.predict(X_test)
print("\n✅ После тюнинга:")
print("Лучшие параметры:", grid.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred_tuned))
print("F1-score:", f1_score(y_test, y_pred_tuned, average='weighted'))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred_tuned))
```

```
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from google.colab import files

# Загрузка файла
uploaded = files.upload()
df = pd.read_excel(next(iter(uploaded)))

# Предобработка данных (аналогично Decision Tree)
df.drop(['Name', 'User_ID'], axis=1, inplace=True)

# Преобразование 'Last_Login' в дату и создание признака 'days_since_login'
df['Last_Login'] = pd.to_datetime(df['Last_Login'], errors='coerce')
df['days_since_login'] = (pd.Timestamp.now() - df['Last_Login']).dt.days
df.drop(['Last_Login'], axis=1, inplace=True)

# Преобразование категориальных переменных в числовые
df['Subscription_Type'] = df['Subscription_Type'].astype('category').cat.codes
df['Favorite_Genre'] = df['Favorite_Genre'].astype('category').cat.codes
df['Country'] = df['Country'].astype('category').cat.codes

# Преобразование 'Watch_Time_Hours' в числовой формат, если есть строки
df['Watch_Time_Hours'] = pd.to_numeric(df['Watch_Time_Hours'], errors='coerce')

# Удаление строк с пропущенными значениями
df.dropna(inplace=True)

# Целевая переменная - heavy_user (пользователи, которые смотрят больше 20 часов)
df['heavy_user'] = df['Watch_Time_Hours'].apply(lambda x: 1 if x > 20 else 0)

# Признаки и цель
X = df.drop('heavy_user', axis=1)
y = df['heavy_user']

# Масштабирование признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение данных на train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# **Step 1: Random Forest Model with Default Hyperparameters**
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf_base = rf_model.predict(X_test)

print("📊 Random Forest (Default Hyperparameters):")
print("Accuracy:", accuracy_score(y_test, y_pred_rf_base))
print("F1-score:", f1_score(y_test, y_pred_rf_base, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_base))

# **Step 2: Hyperparameter Tuning using GridSearchCV**

# Сетка гиперпараметров для Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200, 500],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'max_features': ['auto', 'sqrt', 'log2']
}

grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

# Лучшая модель после настройки
best_rf = grid_search_rf.best_estimator_
y_pred_rf_tuned = best_rf.predict(X_test)

print("\n✅ Random Forest (After Hyperparameter Tuning):")
print("Лучшие параметры:", grid_search_rf.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred_rf_tuned))
print("F1-score:", f1_score(y_test, y_pred_rf_tuned, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf_tuned))

# **Step 3: Comparison with Decision Tree Model**

# Базовая модель Decision Tree
base_model_dt = DecisionTreeClassifier(random_state=42)
base_model_dt.fit(X_train, y_train)
y_pred_dt_base = base_model_dt.predict(X_test)

print("\n📊 Decision Tree (Base Model):")
print("Accuracy:", accuracy_score(y_test, y_pred_dt_base))
print("F1-score:", f1_score(y_test, y_pred_dt_base, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt_base))

# **Compare the performance of Decision Tree and Random Forest**
print("\n🎯 Performance Comparison:")
print("Random Forest Accuracy (Default):", accuracy_score(y_test, y_pred_rf_base))
print("Random Forest Accuracy (Tuned):", accuracy_score(y_test, y_pred_rf_tuned))
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt_base))
```
